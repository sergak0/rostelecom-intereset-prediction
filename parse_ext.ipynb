```python
time_series_covid19_confirmed_global = pd.read_csv("https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
time_series_covid19_confirmed_global = time_series_covid19_confirmed_global[time_series_covid19_confirmed_global["Country/Region"] == "Russia"].T.iloc[4:].reset_index()
time_series_covid19_confirmed_global.columns = ["date", "value"]
time_series_covid19_confirmed_global["date"] = pd.to_datetime(time_series_covid19_confirmed_global["date"])

time_series_covid19_deaths_global = pd.read_csv("https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
time_series_covid19_deaths_global = time_series_covid19_deaths_global[time_series_covid19_deaths_global["Country/Region"] == "Russia"].T.iloc[4:].reset_index()
time_series_covid19_deaths_global.columns = ["date", "value"]
time_series_covid19_deaths_global["date"] = pd.to_datetime(time_series_covid19_deaths_global["date"])
```

```python
def df_covid_convert(df, suffix="", agg=["mean", "median"]):
    df = df.groupby([(df.date.dt.year), (df.date.dt.month)])["value"].agg(agg)
    df.index.names = ("year", "month")
    df = df.reset_index()
    df.columns += suffix
    return df
```

```python
for i in range(1, 7):
  exec(f't{i} = df_covid_convert(google_trends{i}, "_google_trends{i}", agg=["mean"])')
```

```python
time_series_covid19_confirmed_global = df_covid_convert(time_series_covid19_confirmed_global, "_covid19_confirmed")
time_series_covid19_deaths_global = df_covid_convert(time_series_covid19_deaths_global, "_covid19_deaths")
```

```python
df = pd.DataFrame({"period": pd.date_range(start=pd.datetime(2020, 1, 1), end=pd.datetime(2022, 9, 1), freq="M")})
df["month"] = df["period"].dt.month
df["year"] = df["period"].dt.year
df = df.drop(["period"], axis=1)
```

```python
df = df.merge(time_series_covid19_confirmed_global,
              left_on=["year", "month"],
              right_on=["year_covid19_confirmed", "month_covid19_confirmed"],
              how="left")

df = df.merge(time_series_covid19_deaths_global,
              left_on=["year", "month"],
              right_on=["year_covid19_deaths", "month_covid19_deaths"],
              how="left")

df = df.drop(["year_covid19_deaths", "month_covid19_deaths",
              "year_covid19_confirmed", "month_covid19_confirmed"], axis=1)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://bdex.ru/ratings/cities-salary/")
soup = BeautifulSoup(response.content)

data = []
for div in soup.find_all("div", {"class": "row reg_row"}):
    data.append((div.find("div", {"class": "reg_name"}).text,
                 div.find("div", {"class": "reg_fo"}).text,
                 int(div.find("span", {"class": "reg_salary"}).text.replace("\xa0", "")[:-1])))
```

```python
import pickle

with open("cities-salary.pkl", "wb") as f:
    pickle.dump(data, f)
```

```python
train = pd.read_csv("train.csv", delimiter=';')
train["period"] = pd.to_datetime(train["period"], format='%Y-%m-%d')
train["month"] = train["period"].dt.month
train["year"] = train["period"].dt.year
train["city_name"] = train["city_name"].str.strip().str.lower()

with open("cities-salary.pkl", "rb") as f:
    data = pickle.load(f)

data_df = pd.DataFrame(data, columns=["city", "region", 'value'])
data_df["city"] = data_df["city"].str.strip().str.lower()
city2value = dict(data_df[["city", "value"]].values)

train["salary"] = train["city_name"].apply(lambda x: city2value.get(x, 20000))
```
